```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
state_labels = le.fit_transform(df['State_Name'])
state_mappings = {index: label for index, label in enumerate(le.classes_)}
state_mappings
```

```python
df['State_Name'] = state_labels
df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
district_labels = le.fit_transform(df['District_Name'])
district_mappings = {index: label for index, label in enumerate(le.classes_)}
district_mappings
```

```python
df['District_Name'] = district_labels
df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
season_labels = le.fit_transform(df['Season'])
season_mappings = {index: label for index, label in enumerate(le.classes_)}
season_mappings
```

```python
df['Season'] = season_labels
df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
crop_labels = le.fit_transform(df['Crop'])
crop_mappings = {index: label for index, label in enumerate(le.classes_)}
crop_mappings
```

```python
df['Crop'] = crop_labels
df
```

```python
import seaborn as sns
correlation = df.corr()
sns.heatmap(correlation, xticklabels=correlation.columns, yticklabels=correlation.columns)
```

```python
x = df.iloc[:, 0:6]
x
```

```python
y = df.iloc[:, 6:7]
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=3)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
y_pred
```

```python
from sklearn.metrics import mean_absolute_error
errors2 = mean_absolute_error(y_test, y_pred)
errors2
```

```python
# RMSE
from sklearn.metrics import mean_squared_error 
from math import sqrt
error = sqrt(mean_squared_error(y_test, y_pred))
error
```
